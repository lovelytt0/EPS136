2.1 Formula for Magnetic Effect of an Induced Dipole
The master curve derived for the induced dipole is described by the following
mathematical formula.
Note that the x axis was chosen with an origin at the dipole (the origin of the red arrow in
the picture) and pointing to magnetic north. Z was chosen positive down from the dipole
so that y is positive to the east. Distance is normalized by the depth of burial, x/d . The
normalized radius vector is l = (r/d). Note that in MATLAB angles like I must be
converted to radians. The predicted magnetic field is also normalized by the quantity
given in the denominator. It obviously has the units of magnetic field.
The coordinate system used in deriving this formula is illustrated in Fig. 2. North is to the
left along positive x, and z is positive downward from the origin located at the position of
the dipole, d below the surface. The induced dipole, Î¼ , is oriented along the direction of
the ambient magnetic field, B0






```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def indipole(maglat, depth, dipmom, xp, yp):
    '''
    indipole(maglat, depth, dipmom, xp, yp)
    is the function to calculate total magnetic field 
    perturbation of an induced dipole
    at a set of points in a horizontal plane above
    
    return predicted magnetic field in nT (gamma)
    maglat = magnetic latitude of observations
    depth  = depth of dipole in meters
    dipmom = dipole moment in A-m^2
    xp     = magnetic x coordinates of observations
    yp     = magnetic y coordinates of observations
    
    '''

    # Convert magnetic latitude to dipole inclination
    dr  = 180 / np.pi 
    inc = np.arctan(2 * np.tan(maglat / dr))
    ci  = np.cos(inc)
    si  = np.sin(inc)

    # Calculate the constant factor
    con = 100 * (dipmom / depth**3)
    # Normalize the observations
    xd  = xp/depth
    yd  = yp/depth
    l   = np.sqrt(xd**2 + yd**2 + 1)
    # Create the master curve 
    mc  = (3 * (ci * xd - si)**2  / l**5 - 1 / l **3 )

    return mc * con
```

```python
def finzer( xd, mc):
    # a function to find the zero crossing     

    n = len(mc)                          # Get length of mc array
    p = mc[1:n-1] * mc[2:n]              # Sequence of products
    ij = np.where(p <= 0)[0]             # Find negative values
    m = len(ij)                          # Determine # of zero crossings
    zx = np.zeros(m)                     # Create array to hold locations
    
    for i in range(m):                   # Loop for every zero crossing
        # Do a linear interpolation to get the zero crossing more precisely
        ii = ij[i]                       # This is the ith zero crossing
        m =(mc[ii+1]-mc[ii]) / (xd [ii+1]-xd[ii])   # Slope of curve
        b = mc[ii]                       # Intercept of curve
        xx = -b / m                      # Fractional location of zero cross
        zx[i] = xd[ii] + xx              # Absolute location of crossing
        
    return zx
```

```python
# PARAMETERS TO CALC MAGNETIC EFFECT OF AN INDUCED DIPOLE

I = 60                           # The only input parameter
dr = np.pi/180                   # Convert degrees to radians
yd = 0                           # We are taking a N-S profile over object
xd = np.arange(-3, 3, 0.1)       # The inverse cube too weak beyond x/d=3
l = np.sqrt(xd**2 + yd**2 + 1)   # Normalized length of radius vector

# EQUATION
mc = (3 * (np.cos(dr * I) * xd - np.sin(dr * I))**2 / l**5 - 1 / l**3)

# MAKE A PLOT
plt.figure(figsize=(9,9))
plt.plot(xd, mc, 'k-', LineWidth = 1.0)
plt.xlabel('Normalized N-S distance')
plt.ylabel('Normalized magnetic response')
plt.title('MASTER CURVE FOR INDUCED DIPOLE AT 60 DEG LATITUDE')
plt.text(2, .1, 'magnetic north') # Use plot data units for position
plt.plot([-3,3],[0,0],'k--')      # Plot an x-axis
plt.plot([0,0],[-.5,2],'k--')     # Plot a y-axis

# DETERMINATION OF EXTREME VALUES
mxmc = max(mc)
mnmc = min(mc)
xxd = xd[np.argmax(mc)]
nxd = xd[np.argmin(mc)]
print(mxmc, mnmc, xxd, nxd)

# DETERMINATION OF ZERO CROSSINGS
zxd = finzer(xd,mc) 
hxd = finzer(xd, mc-mxmc/2)    # Subtract half of the max
hawid = hxd[1]-hxd[0]          # Half width is distance between zeros
plt.text(0.5,0.4, 'zxd = %5.3f' %zxd, fontsize = 12)
plt.text(0.5,0.25, 'hawid = %5.3f' %hawid, fontsize = 12)

plt.savefig('Lab1.png')
```
